```python
dfs = []

# Loop through the CSV files and read them into DataFrames
for year in range(2000, 2011):
    filename = f'wta_matches_{year}.csv'
    df = pd.read_csv(filename)
    unique_tourney_ids = df['tourney_id'].nunique()
    result_df = pd.DataFrame({'Year': [year], 'Number of Tournaments': [unique_tourney_ids]})
    dfs.append(result_df)

# Concatenate the list of DataFrames into one
result_df = pd.concat(dfs, ignore_index=True)

# Create a bar chart using Plotly Express
# fig = px.bar(result_df, x='Year', y='Number of Tournaments', title='Number of Tournaments by Year')
fig = px.bar(result_df, x='Number of Tournaments', y='Year', orientation='h', 
             title='Number of Tournaments by Year', labels={'Year': 'Year'})
fig.show()
fig.write_image('viz_2.png')
```

```python
import pandas as pd
import plotly.graph_objs as go
df= pd.read_csv('wta_players.csv',low_memory=False)

# Load custom human icon image
```

```python
tennis_df = pd.read_csv('wta_matches_1970.csv')

left_handed_winners = tennis_df[tennis_df['winner_hand'] == 'L']

# Filter for rows where 'winner_hand' is 'R' (right-handed)
right_handed_winners = tennis_df[tennis_df['winner_hand'] == 'R']

# Create bins for age groups
age_bins = list(range(0, 100, 10))

# Calculate the count of winners in each age group for both hands
left_handed_counts = []
right_handed_counts = []

for bin_start, bin_end in zip(age_bins[:-1], age_bins[1:]):
    left_handed_count = len(left_handed_winners[(left_handed_winners['winner_age'] >= bin_start) & (left_handed_winners['winner_age'] < bin_end)])
    right_handed_count = len(right_handed_winners[(right_handed_winners['winner_age'] >= bin_start) & (right_handed_winners['winner_age'] < bin_end)])
    
    left_handed_counts.append(left_handed_count)
    right_handed_counts.append(right_handed_count)

# Reverse the right-handed counts for the population pyramid
right_handed_counts.reverse()

layout = go.Layout(yaxis=go.layout.YAxis(title='Age'),
                   xaxis=go.layout.XAxis(
                       range=[-max(left_handed_counts + right_handed_counts), max(left_handed_counts + right_handed_counts)],
                       tickvals=[-max(left_handed_counts), 0, max(left_handed_counts)],
                       ticktext=[max(left_handed_counts), 0, max(left_handed_counts)],
                       title='Number'),
                   barmode='overlay',
                   bargap=0.1)

data = [go.Bar(y=age_bins[:-1],
               x=left_handed_counts,
               orientation='h',
               name='Left-Handed Winners',
               hoverinfo='x',
               marker=dict(color='powderblue')
               ),
        go.Bar(y=age_bins[:-1],
               x=right_handed_counts,
               orientation='h',
               name='Right-Handed Winners',
               text=-1 * right_handed_counts,
               hoverinfo='text',
               marker=dict(color='seagreen')
               )]

# Create the population pyramid
fig = go.Figure(data=data, layout=layout)

# Show the interactive plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px
tennis_df = pd.read_csv('wta_matches_2022.csv')

# tourney name: "BJK Cup QLS R1: USA vs UKR"
tournament_ids_to_match = [
    '2022-W-FC-2022-QLS-M-USA-UKR-01'
]

tennis_df = tennis_df[tennis_df['tourney_id'].isin(tournament_ids_to_match)]

selected_columns = ['w_ace', 'w_df', 'w_svpt', 'w_bpSaved', 'w_bpFaced']
tennis_df = tennis_df[['winner_name'] + selected_columns].dropna()

for column in selected_columns:
    max_value = tennis_df[column].max()
    tennis_df[column] = 10* (tennis_df[column] / max_value)


players = tennis_df['winner_name'].unique()


data = []

for player in players:
    player_data = tennis_df[tennis_df['winner_name'] == player][selected_columns].values[0].tolist()
    player_data += [player_data[0]] 
    data.append(go.Scatterpolar(
        r=player_data,
        theta=selected_columns + [selected_columns[0]],
        fill='toself',
        name=player
    ))

layout = go.Layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
        ),
    ),
    showlegend=True,
    title='Radar Chart of winners performance in BJK Cup QLS R1: USA vs UKR in 2022',
)

fig = go.Figure(data=data, layout=layout)

fig.show()
fig.write_image('radar-chart.png')
```

```python
df = pd.read_csv('wta_matches_2018.csv')
df = df.dropna(subset=['minutes'])
independent_var1 = 'tourney_level'
independent_var2 = 'surface'
dependent_var = 'minutes'


pivot_df = df.pivot_table(index=independent_var2, columns=independent_var1, values=dependent_var, aggfunc='mean')
pivot_df = pivot_df.fillna(0)


fig = px.imshow(pivot_df,
                labels=dict(x=independent_var1, y=independent_var2, color=dependent_var),
                x=pivot_df.columns,
                y=pivot_df.index,
                color_continuous_scale='Viridis')


fig.update_layout(title=f"Heatmap of {dependent_var} by {independent_var2} and {independent_var1} in 2018")


for i in range(len(pivot_df.index)):
    for j in range(len(pivot_df.columns)):
        fig.add_annotation(
            text=str(round(pivot_df.iloc[i, j], 2)),  # Display rounded values
            x=pivot_df.columns[j],
            y=pivot_df.index[i],
            showarrow=False,
            font=dict(size=12)
        )


fig.show()
fig.write_image('heatmap.png')

```
